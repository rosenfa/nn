This Notebook will introduce the basics of gradient descent.  We will use the function, x*x*np.log(x)-x and derivative 2*x*np.log(x)+x-1 we discussed in class.

```python
import matplotlib.pyplot as plt
import numpy as np
#cost = y - y-hat (guess)
def function(x):
    return x*x*np.log(x)-x

def function_prime(x):
    return 2*x*np.log(x)+x-1
```

Let's take the values of the function in steps of 0.1 (using numpy's arange method) for both the function and the derivative.

```python
x1 = np.arange(0.1,5,0.1)
y1= function(x1)
    
x2 = np.arange(0.1,5,0.1)
y2= function_prime(x1)
```

Let's plot them now...

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.grid()

plt.xlim(0,4)

plt.title('Functions in Numpy!',fontsize=10)

plt.xlabel('X',fontsize=14)
plt.ylabel('Y',fontsize=14)

plt.savefig('how_to_caclulate_natural_logarithm_in_python.png', bbox_inches='tight')
plt.show()
```

Here is a basic gradient descent implementation.  We start the search for the minimum at 2 (we know it's 1, but this shows how it can be experically found).  Every iteration we print t, the derivative of the function we are trying to solve, and update the value, next_x. We end the loop when we hit max_iter (max. iterations) or when we break out because the difference, difference, between the old and new value is less than epsilon (0.00001).  Feel free to play with the values for alpha -- the step size. 

```python
next_x = 2
alpha = 0.2
max_iter=50
epsilon = 0.00001
print("The new initial value of the search is " + str(next_x))
for i in range(max_iter):
    current = next_x
    t=function_prime(current)
    next_x = next_x - alpha*t
    print("Iteration number: " + str(i+1))
    print("The derivative is " + str(t))
    print("The change in value is " + str(alpha*t))
    print("The new value in the search is " + str(next_x))
    difference = next_x - current
    #print(difference)
    if (abs(difference)<epsilon):
        break
print("answer is " + str(next_x))
```

Sometimes have a large learning rate will prevent us from converging.  Let's consider two functions: x^2 âˆ’ 2x + 3 and its derivative, 2x - 2. Note that the optimal minimum will be at x = 1 (2x - 2 = 0), but let's start the search at 2 and try to find that out with gradient descent with alpha set to 20. 

```python
next_x = 2
alpha = 20 #instead of 0.2
max_iter=50
epsilon = 0.00001
print("The new initial value of the search is " + str(next_x))
for i in range(max_iter):
    current = next_x
    t=(2 * next_x) - 2 #function_prime(current)
    next_x = next_x - alpha*t
    print("Iteration number: " + str(i+1))
    print("The derivative is " + str(t))
    print("The change in value is " + str(alpha*t))
    print("The new value in the search is " + str(next_x))
    difference = next_x - current
    #print(difference)
    if (abs(difference)<epsilon):
        break
print("answer is " + str(next_x))
```
